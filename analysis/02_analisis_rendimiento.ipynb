# Análisis del Rendimiento

En este archivo hablaremos sobre el método de validación elegido y el método de evaluación utilizado.

### Método de validación

Tenemos varios métodos de validación posibles:
- Hold Out (estratificación y repetición)
- K-Fold (estratificación)
- Leave-One-Out

Nuestro dataset tiene 704 ejemplos, por lo que Hold-Out no la usaremos porque el número de veces que
se usaría cada ejemplo para entrenar el modelo podría ser bastante diferente, y podría crear un sesgo.
Como consecuencia, solo nos quedaría el K-Fold y el Leave-One-Out, sin embargo, si usáramos el Leave-One-Out
obligaríamos a entrenar nuestra pipeline unas 700 veces, lo cual es bastante costoso, por lo tanto usaremos K-Fold.

Para aplicar el K-Fold necesitamos saber el valor de K, el cual suele ser 5 o 10. Si eligieramos 5,
estaríamos entrenando con el 80% de los datos que nos llegaran y si fuera 10, usaríamos el 90%, así
que como tenemos pocos ejemplos, es preferíble K=10.

Asimismo, para busqueda de los mejores hiperparámetros tenemos 2 técnicas:
- GridSearch
- RandomSearch

GridSearch utiliza la búsqueda exhaustiva de todos los valores que se le introducen, por lo que si hay
muchos hiperparámetros con espacios de búsqueda muy extensos se necesitarían demasiadas iteraciones para
cubrir el espacio posible, sin embargo, una solución sería probar con muchos menos valores fijos, pero
se podría introducir un sesgo. En conclusión, GridSearch se usa cuando hay pocos hiperparámetros con
espacios de búsqueda pequeños.

Por otro lado, tenemos el RandomSearch que sirve para justo lo contrario, es decir, como los valores
de los hiperparámetros se eligen de forma aleatoria siguiendo una distribución de probabilidad, vienen
muy bien cuando hay muchos hiperparámetros con espacios de búsqueda extensos.

Para nuestro problema usaremos RandomSearch para que nuestra busqueda de hiperparámetros sea más flexible
y no esté sesgada.

### Método de evaluación

Tenemos varios métodos de evaluación posibles:
- Dependientes del umbral:
    - Accuracy
    - F1-score

- Independientes del umbral:
    - AUC de la curva ROC
    - AUC de la curva Precisión-Recall

Para que nuestra medida sea bastante flexible deberíamos de usar las medidas independientes del umbral,
por lo que todas las medidas dependientes del umbral quedan descartadas. Ahora toca ver si quedarnos
con el AUC de la curva ROC o el AUC de la curva Precisión-Recall.

En cuanto a la curva ROC, esta se suele utilizar para comparar modelos entre sí, y la curva Precisión-Recall
para elegir el mejor umbral de nuestro modelo en función de si queremos mayor precisión o mayor recall.
Por tanto, la medida que utilizaremos para nuestra pipeline será el AUC de la curva ROC.
